```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(536)
X = np.linspace(0, 200, 20)
Y = np.random.randint(100, size=20)
plt.figure()
plt.plot(X, Y)
```

On a une fonction discrete et on cherche les nombres derivés pour chaque x
On cherche a trouver les derivés nul

```python
derivatives = []
old_derivatives = []
nul_derivatives = []
p_derivatives = []
m_derivatives = []
x_nul = []
y_nul = []

for i, x in enumerate(X[:-1]):
    delta_Y = Y[i + 1] - Y[i]
    delta_X = X[i + 1] - x
    
    
    derivative = delta_Y / delta_X
    
    derivatives.append(int(derivative))
    old_derivatives.append(derivative)
    
    if int(derivative) == 0:
        nul_derivatives.append(i)
    elif int(derivative) >= 4:
        p_derivatives.append(i)
    elif int(derivative) <= -4:
        m_derivatives.append(i)

for place in nul_derivatives:
    x_nul.append(X[place])
    y_nul.append(Y[place])
    
```

```python
np.random.seed(536)

plt.figure()
plt.plot(X, Y)
plt.scatter(x_nul, y_nul, c="red", marker = 'o')
```
