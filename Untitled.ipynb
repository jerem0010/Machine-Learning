```python
import numpy as np
```

```python
x = np.linspace(0, 2, 10)
y= x**2
print()
```

1) Pyplot

```python
import matplotlib.pyplot as plt
```

Y en fonction de X

```python
plt.plot(x, y, c="yellow")
plt.show()
```

```python
plt.scatter(x, y)
plt.show()
```

Mis en place de 2 courbes sur le meme graphique

```python
plt.figure()
plt.plot(x, y, label='quadratique')
plt.plot(x, x**3, label='cubique')
plt.title('Figure 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
```
